```python
#reading csv data
df_train=pd.read_csv('Training.csv')
```

## Simple Exploratory Data Analysis

### Converting prognosis column into numerical

### Label Encoding

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
df_train['prognosis'] = le.fit_transform(df_train['prognosis'])
```

```python
df_train
```

### Performing EDA

### Analysing the bar plots

```python
colors='#3d84bf','#295981'
columns=list(df_train.columns)
#barplot of the count for all symptoms absence and presence
for i in columns:
    fig,ax= plt.subplots(figsize=(2,2))
    bar=df_train.groupby(i).size().plot(kind='bar',color=colors,ax=ax)
    plt.xticks(rotation=0)
    fig.suptitle("count of symptom \"" + i + "\"")
```

# splitting the data

```python
X = df_train.drop(columns=['prognosis'])
y = df_train['prognosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44, shuffle =True)
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogisticRegressionModel = LogisticRegression()
LogisticRegressionModel.fit(X_train, y_train)

print('LogisticRegressionModel Train Score is : ' , LogisticRegressionModel.score(X_train, y_train))
print('LogisticRegressionModel Test Score is : ' , LogisticRegressionModel.score(X_test, y_test))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
y_pred = LogisticRegressionModel.predict(X_test)
CM_LR = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is\n', CM_LR)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dtc = DecisionTreeClassifier(criterion='entropy',max_depth=23,random_state=44)
Dtc.fit(X_train, y_train)
print('DecisionTreeClassifierModel Train Score is : ' , Dtc.score(X_train, y_train))
print('DecisionTreeClassifierModel Test Score is : ' , Dtc.score(X_test, y_test))
```
